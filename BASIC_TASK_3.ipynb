# **TASK 3: Create a Basic Classification Model**

```python
# Step 1: Import libraries
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/iris.csv')
df
```

```python
# Step 3: Separate features and target
X = df.iloc[:, :-1]   # All columns except last one (features)
y = df.iloc[:, -1]    # Last column (species)
```

```python
# Step 4: Train the Logistic Regression model
model = LogisticRegression(max_iter=200)
model.fit(X, y)
```

```python
# Step 5: Make predictions
predictions = model.predict(X)
```

```python
# Step 6: Combine Actual and Predicted into a DataFrame
results = pd.DataFrame({
    "Index": range(len(y)),
    "Actual Label": y,
    "Predicted Label": predictions
})
```

```python
# Step 7: Display first 10 rows only for clarity
print(results.head(60).to_string(index=False))
```

**Train the Decision Tree classification model**

```python
model = DecisionTreeClassifier()
model.fit(X, y)
```

```python
predictions = model.predict(X)
results = pd.DataFrame({
    "Index": range(len(y)),
    "Actual Label": y,
    "Predicted Label": predictions
})
```
